**CREACIÓN DE CENTROIDES**
Existe variedad de formatos que pueden ser utilizados para representar los elementos geoespaciales ya sea puntos, líneas y polígonos. Geopandas permite la lectura de datos espaciales vectoriales en formatos como shapefile, geojson y geopackage.

```python

ANP_SRC["centroide"] = ANP_SRC.centroid
ANP_SRC
```

```python
punto_inicial = ANP_SRC["centroide"].iloc[0]
```

```python
punto_inicial
```

```python
ANP_SRC["distancia"] = ANP_SRC["centroide"].distance(punto_inicial)
ANP_SRC["distancia"]
```

```python
ANP_SRC["distancia"].mean()
```

```python
ANP_SRC.plot("area", legend = True)
```

```python
ANP_SRC = ANP_SRC.set_geometry("centroide")
```

```python
ANP_SRC.plot("area", legend = True)
```

```python
#Colocar ambos gráficos
axis = ANP_SRC["geometry"].plot()
ANP_SRC["centroide"].plot(ax = axis, color = "black")
```

```python
ANP_SRC = ANP_SRC.set_geometry("geometry")
ANP_SRC.plot(column = "anp_cate", legend = True, legend_kwds = {"loc": "upper right"}, figsize = (20, 12))
```

```python
ANP_SRC["envolvente"] = ANP_SRC.convex_hull
```

```python
axis = ANP_SRC["envolvente"].plot(alpha = 0.5)
ANP_SRC["perimetro"].plot(ax = axis, color = "red", linewidth = 0.2)
```

```python
ANP_SRC["buffer"] = ANP_SRC.buffer(10000)
```

```python
ANP_SRC["buffer_centroide"] = ANP_SRC["centroide"].buffer(10000)
```

```python
axis = ANP_SRC["buffer"].plot(alpha = 0.5)
ANP_SRC["buffer_centroide"].plot(ax = axis, color = "red", alpha = 0.5)
ANP_SRC["perimetro"].plot(ax = axis, color = "white", linewidth = 0.5)
```

```python
#Método de pandas que realiza la extracción de una etiqueta
ANP_SRC = ANP_SRC.set_geometry("geometry")
Tabaconas = ANP_SRC.loc["Tabaconas-Namballe", "geometry"]
Tabaconas
```

```python
ANP_SRC["buffer"].intersects(Tabaconas)
```
